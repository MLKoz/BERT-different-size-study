<h3>Generalne informacje o treningu i wstępnym przetwarzaniu danych:</h3>
<ol>

<li>LSTM: Współczynnik uczenia 1e-04, algorytm uczenia Adam, sposób treningu - from scratch, batch size 16. Architektura: 3 warstwy po 192 komórki LSTM każda. </li>
<li>Wykorzystane GPU: NVIDIA TESLA P100  16GB, czas obliczeń 200h.</li>

</ol>

<center><h2>Wnioski</h2></center>
<ol>
  <li>BERT Tiny uzyskał średnio wyższe wyniki (F1 - weighted) niż LSTM na co wpływ może mieć fakt, że wraz z BERT Tiny wykorzystano już wstępnie wytrenowany model (fine-tuning), a LSTM było trenowane od podstaw. Dodatkowo BERT z tą samą liczbą parametrów/wag do wyuczenia wydaje się lepiej obsługiwać dłuższe sekwencję - prawdopodobnie dzięki mechanizmowi atencji.</li>
  <li>Wykorzystując fine-tuning w BERT z tą samą liczbą parametrów co LSTM trenowana od podstaw, trenuje się szybciej, istnieje opcja by zrównoleglić obliczenia oraz zazwyczaj daje lepsze wyniki (dokładność, wskaźnik F1, precyzja, recall). Mechanizm atencji wydaje się dawać lepsze rezultaty niż komórki pamięci LSTM.</li>
  <li>Im większy model BERT tym zazwyczaj lepsze wyniki, lecz wyraźną różnicę widać przy dłuższych sekwencjach - standardowy model BERT oraz wersja large dają lepsze wyniki dla dłuższych sekwencji. Niestety wykorzystanie dużych modeli (normal 24 razy większy od tiny, oraz large 75 razy większy od tiny) wymaga wielokrotnie dłuższych obliczeń oraz pamięci GPU - co może być czasem nieosiągalne do zapewnienia. Do treningu wykorzystano kartę z 16GB pamięci co jest drogim kosztem, czas obliczeń to około 200h, pomimo takiej ilości pamięci, niektóre zbiory danych nie mieściły się pomimo użycia małego batch_size.</li>
  <li>Nie zawsze użycie dużo większego modelu dawało lepsze rezultaty , szczególnie dla krótkich sekwencji - prawdopodobnie przez wzrost ryzyka nadmiernego dopasowania.</li>
  <li>Często ważniejsza od wielkości modelu była długość przetwarzanej sekwencji, najmniejszy model BERT często dla długich sekwencji liczył się dość szybko oraz dawał dobre rezultaty jednocześnie przy małym zapotrzebowaniu na pamięć GPU.</li>
  <li>Zwiększenie rozmiarów glove wydaje się nie mieć wpływu na jakość działania LSTM - prawdpodobnie trzeba rozmiar glove ustalić jako hiperparametr argumentu i do każdego zbioru/wielkości modelu dobrać glove indywidualnie.</li>
</ol>
<center><h2>Możliwe dalsze kierunki</h2></center>
<ol>
  <li>Porównanie innej sieci LSTM np: CNN-LSTM, dwukierunkowa sieć LSTM, sieć LSTM z mechanizmem atencji.</li>
  <li>Wykorzystanie do badań mechanizmu destylacji wiedzy i porównanie modeli BERT poddanych destylacji (a w szczególności BERT Tiny) z normalnymi oraz LSTM zwykłą o tej samej liczbie parametrów/destylowanej.</li>
  <li>Wykorzystanie zespołu modeli do uzyskania lepszych wyników, np. różne rodzaje BERT + LSTM w celu osiągnięcia wyników state-of-the-art.</li>
  <li>Wykorzystanie innej wersji metody klasyfikacji niż tej opierającej się na tokenie CLS - np. average (uśrednianie) lub pooling/max.pooling.</li>
</ol>
